```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
from scipy.integrate import quad
from scipy.stats import norm

mu = 200
variance = 2
sigma = math.sqrt(variance)
frozen = norm(mu, sigma)
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)
plt.plot(x, frozen.pdf(x))
plt.show()
```

```python
first_point = quad(frozen.pdf, 205, np.inf)[0]
second_point = quad(frozen.pdf, 199, 201)[0]
third_point = quad(frozen.pdf, 203, np.inf)[0] * quad(frozen.pdf, 203, np.inf)[0]
```

```python
mu = 0
sigma = 1
frozen = norm(mu, sigma)
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)
plt.plot(x, frozen.pdf(x))
plt.show()
```

```python
cdf = [quad(frozen.pdf, -np.inf, xi)[0] for xi in x]
```

```python
plt.plot(x, cdf)
plt.show()
```

```python
f = lambda x, mu=0, sigma=1 : norm.pdf(x, mu, sigma)
values = np.linspace(-5,5,100)
N = 10**6
x = np.random.uniform(-5, 5, N)
y = np.random.uniform(0.0, 1.0, N)
random_hit_miss = x[(y < f(x))]
fig, ax = plt.subplots()
ax.hist(random_hit_miss, bins=np.arange(-5, 5 + 0.25, 0.25), density=True, label = 'Hit or Miss') 
ax.plot(values, f(values), 'r', label='Gauss $\mu=0, \sigma=1$')
ax.legend();
```
