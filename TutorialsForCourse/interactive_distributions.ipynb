```python
%matplotlib widget
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()

@widgets.interact(l=(500, 2000, 100))
def update(l=500):
    size = 1000000
    [l.remove() for l in ax.containers]
    sample = np.random.poisson(l, size)
    ax.set_xlim(250, 2250)
    ax.set_ylim(0, 0.02)
    ax.hist(sample, bins=100, color='C0', density=True)
```
